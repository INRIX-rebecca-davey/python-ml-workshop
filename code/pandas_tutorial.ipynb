```python
# taking a subset of the columns

df = dfRawEnc[['Reference Number', 'Latitude', 'Longitude', 'Age of Casualty']]
df.head(5)
```

```python
# renaming columns

df.columns = ['referenceNumber', 'latitude', 'longitude', 'age']
df.head(5)
```

```python
# setting a whole new column to a fixed value

df['fixedColumn'] = 3
df.head(5)
```

```python
# combining existing columns

df['latPlusLng'] = df['latitude'] + df['longitude']
df.head(5)
```

```python
# writing a function to combine existing columns in a more complex way

# if the first digit of the reference number is even then return the latitude squared, otherwise the longitude squared
def complexFunction(row):
    refFirstDigit = int(str(row['referenceNumber'])[0])
    
    if refFirstDigit % 2 == 0:
        latSq = row['latitude'] * row['latitude']
        return latSq
    else:
        lngSq = row['longitude'] * row['longitude']
        return lngSq
    
df['latOrLngSq'] = df.apply(complexFunction, axis=1)
df.head(5)
```

```python
# filtering the rows

under30 = df[df.age < 30]
under30.head(5)
```

```python
# dropping columns

noLat = df.drop(columns=['latitude'])
noLat.head(5)
```

```python
# saving data

outPath = "pandas-tutorial-output.csv"
df.to_csv(outPath, index=False, header=True) # will save as csv without index column but with column names as header
```
